# Plot metatopic sharing distribution by outlet (Fig. 4)

```python
print("Loading data...")
full_df = pd.read_csv("./data/articles_to_id.csv") ## replace by your path to data
print("Done")
```

```python
print("Loading doc-top-matrix...")
dtm = pd.read_csv(f"./data/article_topic_matrix.csv")
```

```python
edgelist = pd.read_csv('./data/shares_pseudo.csv', dtype = str)
edgelist = edgelist.drop_duplicates(subset = ['tweet_id'])

edgelist = edgelist.merge(full_df, left_on = 'url', right_on = 'link')

edgelist = edgelist.merge(dtm, left_on = 'article_id', right_on = 'article_id')
```

```python
party_positions = pd.read_csv('./data/party_pos_FINAL.csv')

user_embedding = './data/embedding_pseudo.csv'

user_emb_df = pd.read_csv(user_embedding, dtype = str)
user_emb_df = user_emb_df.set_index('user')

user_list = list(edgelist.user)
pos_dict = user_emb_df.T.to_dict('list')
```

```python
pos_list_x = []
pos_list_y = []

ct = 0
ne_ct=0

for user_id in tqdm(user_list):
    if user_id in pos_dict.keys():
        pos_list_x.append(pos_dict[user_id][0])
        pos_list_y.append(pos_dict[user_id][1])
        ct +=1
    else:
        pos_list_x.append('not_embedded')
        pos_list_y.append('not_embedded')
        ne_ct +=1

edgelist['x'] = pos_list_x
edgelist['y'] = pos_list_y
```

```python
meta_df = pd.read_csv('./data/assigned_metatopics.csv') 
```

```python
def plot_metatopic_distribution(outlet, metatopic, edgelist, party_positions, meta_df, follower_color, follower_alpha, hist_alpha, density=True):  
    
    outlet_df = edgelist[edgelist['outlet_x'] == outlet]
    
    title = outlet + ': ' + metatopic

    id_df = meta_df[meta_df['metatopic'] == metatopic]
    topic_list = list(id_df.topic)

    x = []
    y = []
    limit = 0.5

    new_topic_list = []
    for element in topic_list:
        top = str('Topic') + str(element)
        new_topic_list.append(top)
        
    outlet_df['summed_p'] = outlet_df[new_topic_list].sum(axis=1)

    frame = outlet_df[outlet_df['summed_p'] >= limit]
    x = [float(i) for i in list(frame['x']) if i != 'not_embedded']
    y = [float(i) for i in list(frame['y']) if i != 'not_embedded']
    #for element in topic_list:
    #    top = str('Topic') + str(element)

    #    frame = outlet_df[outlet_df[top] >= limit]
    #    x1 = [i for i in list(frame['x']) if i != 'not_embedded']
    #    y1 = [i for i in list(frame['y']) if i != 'not_embedded']
    #    x = x + x1
    #    y = y + y1

    x = [float(s) for s in x]
    y = [float(s) for s in y]
    
    FOLLOWER_COLOR = follower_color
    FOLLOWER_ALPHA = follower_alpha
    STORY_COLOR_MEAN = "red"    

    mean_x = sum(x)/len(x)
    mean_y = sum(y)/len(y)    
    party_positions['0'] = party_positions['0'].astype(float)
    party_positions['2'] = party_positions['2'].astype(float)
    x_mp = list(party_positions['0'])
    y_mp = list(party_positions['2'])
    
    fig = plt.figure(figsize=(10,10))
    
    gs = fig.add_gridspec(2, 2,  width_ratios=(7, 1), height_ratios=(1, 7),
                          left=0.1, right=0.9, bottom=0.1, top=0.9,
                          wspace=0.005, hspace=0.005)
    
    ## define axes
    ax = fig.add_subplot(gs[1, 0])
    ax_histx = fig.add_subplot(gs[0, 0], sharex=ax)
    ax_histy = fig.add_subplot(gs[1, 1], sharey=ax)    
    
    if density == True:
        plot = sns.kdeplot(x=x, y=y, levels=20, cmap="Blues", 
                    fill = False,alpha = 1, zorder = 2, ax=ax)
        
        sharer_patch = Line2D([0], [0], color=follower_color, 
                       markerfacecolor=follower_color)
        
    else:
        plot = ax.scatter(x=x, y=y, c=FOLLOWER_COLOR, 
                    alpha = .1, zorder = 2, marker = 'x')
    #smean = ax.scatter(mean_x, mean_y, c = 'red', marker = 'x', s = 150, zorder = 3)
    colorlist = list(party_positions['color'])
    party_positions.plot(kind='scatter', x = '0', y = '2', c = colorlist, zorder = 3, marker = 'o', ax=ax, edgecolors='black', s=130, alpha = .7)    
    
#     ax.set_title(outlet + ': Outlet sharing distribution')
    
    ## title and axis labels
    ax.set_title(title, 
                       fontsize = 20,
                       loc='left',
                       fontdict = {'weight':'bold'},
                       x=0.02, y=0.93)    
    
    ax.set_xlabel('Left-right', fontsize = 17)
    ax.set_ylabel('Elite-/EU-skeptical/Protectionist', fontsize = 17)
    ax.set_ylim(-2,4)
    ax.set_xlim(-3,3)    
    
    ## define and draw histograms
    binwidth = 0.05
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth
    bins = np.arange(-lim, lim + binwidth, binwidth)    
    ax_histx.hist(x, bins=bins, density=False,color=FOLLOWER_COLOR,alpha=hist_alpha)
    ax_histy.hist(y, bins=bins, orientation='horizontal',color=FOLLOWER_COLOR,alpha=hist_alpha)                
    
    ## some details
    ax_histx.axis('off')
    ax_histy.axis('off')
    ax.grid(alpha=.3)
    ax.tick_params(direction='in',which='both',width=1,top=True,right=True)
    ax.tick_params(direction='in',which='major',width=1,length=10)
    ax.tick_params(direction='in',which='minor',width=1,length=5)    
    
    ## add a custom legend element for the party positions
    patch = Line2D([0], [0], marker='o', color='w', 
                   markerfacecolor='w', markeredgecolor='black', markersize=7)
    
    ## define legend
    if density == True:
        legend = ax.legend([sharer_patch, patch],
                           ('Distribution of shares', 'Party position'),
                           loc=(0.03, 0.05),
                          )
    else:
        legend = ax.legend([plot, patch],
                           ('Outlet sharers', 'Party position'),
                           loc=(0.03, 0.05),
                          )        
    
    # make sure all legend markers have the same size
    for marker_idx in range(2):
        legend.legend_handles[marker_idx]._sizes = [50]
    for i in range(7):
        if i != 3:
            ax.get_xticklabels()[i].set_color("white")
    for i in range(7):
        if i != 2:
            ax.get_yticklabels()[i].set_color("white")
    
    name = './' + str(outlet) + str(metatopic).replace('/','') + '_added.pdf'
    plt.savefig(name, dpi = 300)
    plt.show()
    plt.close()
    
    return None    
```

```python
plot_metatopic_distribution('welt.de', 'environment/climate change',
                    edgelist,
                    party_positions,
                    meta_df,
                    follower_color='#6495ED',
                    follower_alpha=.4,
                    hist_alpha=.8,
                    density=True)
```
